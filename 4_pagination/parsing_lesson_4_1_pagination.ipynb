```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
def get_html(url):
    r = requests.get(url)
    if r.ok: #200 ok ## 403 404 ne_ok
        #print(r.status_code)
        return r.text
    #print(r.status_code)
        
```

```python
def write_csv(data):
    with open('yand.csv', 'a', newline = '\n') as f:
        writer = csv.writer(f)
        writer.writerow([data['name'],
                         data['link']])
```

```python
def get_page_data(html):
    #encodedText = r.text.encode("utf-8")
    soup = BeautifulSoup(html, 'lxml')
    lis = soup.find_all('li', class_ = 'serp-item')
    
    #print(soup)
    #print(len(lis))
    
    for li in lis:
        try:
            name = li.find('div', class_ = 'organic__url-text').text
        except:
            name = ''
            
        try:
            #link = li.find('a', class_ = 'link link_theme_outer path__item i-bem link_js_inited').text
            link = li.find('a', class_ = "link link_theme_normal organic__url link_cropped_no i-bem").get('href')
        except:
            link = ''
        
        data = {'name': name,
                'link': link}
        write_csv(data)
            
        print(name, link, sep = ' ')
```

```python
def main():
    url = 'https://yandex.ru/search/?lr=213&text=%D1%81%D0%BF%D0%B8%D1%81%D0%BE%D0%BA%20%D1%81%D0%B0%D0%B9%D1%82%D0%BE%D0%B2&p={}'
    for i in range(0,5):
        get_page_data(get_html(url.format(str(i))))
```

```python
url = 'https://yandex.ru/search/?lr=213&text=%D1%81%D0%BF%D0%B8%D1%81%D0%BE%D0%BA%20%D1%81%D0%B0%D0%B9%D1%82%D0%BE%D0%B2&p=0'
b_soup = get_html(url)
```
